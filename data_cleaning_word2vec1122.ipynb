```python
import pandas as pd
import spacy #spacy is a natural language processing library used here for lemmatization and working with English text.

# Load spaCy English model
nlp = spacy.load("en_core_web_sm")

# Load the CSV file
df = pd.read_csv("/content/drive/MyDrive/FYP/preprocessed_transcriptions.csv")  # Use your actual file path

# Emergency keyword definitions (nouns and verbs)
emergency_keywords = {
    'Accident_Call': {
        'accident', 'crash', 'collision', 'hit', 'strike', 'vehicle', 'wreck',
        'overturn', 'injured', 'bump', 'pileup', 'unconscious', 'motorcycle',
        'car crash', 'road accident', 'bike accident', 'truck collision', 'pedestrian hit','blood'

    },
    'Fire_Call': {
        'fire', 'burn', 'flame', 'blaze', 'ignite', 'smoke', 'cylinder', 'explosion',
        'wildfire', 'combustion', 'scorch', 'ash', 'inferno', 'char', 'sparks', 'extinguish', 'alarm',
        'gas leak', 'short circuit', 'electric fire', 'blast', 'firefighter'

    },
    'Medical_Call': {
        'kidney', 'chest', 'abdomen', 'liver', 'heart', 'lung', 'brain', 'stomach', 'appendix',
        'injury', 'unconscious', 'pain', 'bleed', 'medical', 'sick', 'hurt', 'hurting','faint', 'suicide', 'patient','blood',
        'seizure', 'heartattack', 'breathing', 'allergy', 'stroke', 'vomit','vomiting', 'fever', 'diabetic', 'headache'

    }
}


# Combine all keywords into one set
all_emergency_keywords = set().union(*emergency_keywords.values())

# Function to extract relevant keywords
def extract_relevant_keywords(lemmatized_text):
    try:
        tokens = eval(lemmatized_text) #eval() is used to convert it into a list.
    except:
        return []
    keywords = [token for token in tokens if token.lower() in all_emergency_keywords]
    return list(dict.fromkeys(keywords))[::]  # remove duplicates, limit to 4

# Apply keyword extraction
df['Keywords'] = df['Lemmatized_Transcription'].apply(extract_relevant_keywords)


# Function to assign emergency label (first match wins)
def assign_emergency_label(keywords):
    for word in keywords:
        word = word.lower()
        for label, keywords_set in emergency_keywords.items():
            if word in keywords_set:
                return label
    return 'Normal_Call'  # if no keywords matched

# Apply label assignment
df['Label'] = df['Keywords'].apply(assign_emergency_label)

# Save the complete DataFrame
df.to_csv("/content/drive/MyDrive/FYP/final_keywords_emergency_index.csv", index=False)

print("✅ Done! Labels assigned simply based on keyword matching.")

```

#**KeyWords Extractions**

```python
import pandas as pd
import ast
from collections import Counter
df = pd.read_csv("/content/drive/MyDrive//FYP/final_keywords_emergency_index.csv")
# Define meaningless/common filler words to exclude
irrelevant_words = set([
    'press', 'one', 'two', 'three', 'four', 'five',
    'ji', 'al', 'alaikum', 'assalam', 'say', 'said',
    'please', 'rescue', 'double', 'call', 'yes', 'no',
    'number', 'mujhy', 'k', 'mein', 'hain', 'ka', 'ki',
    'ke', 'ho', 'raha', 'gaya', 'kar', 'tha', 'hai', 'the',
    'please', 'aap', 'ko', 'se', 'for', 'in', 'on', 'is'
])

# (Optional) Emergency-relevant vocabulary to prioritize (could be expanded)
emergency_keywords_whitelist = set([
    'accident', 'injury', 'bleeding', 'fire', 'cylinder', 'burn',
    'unconscious', 'patient', 'blood', 'gas', 'pain', 'death',
    'fainted', 'ambulance', 'vehicle', 'hit', 'truck', 'motorcycle'
])

# Function to extract meaningful emergency-related words
def extract_emergency_related_words(row):
    try:
        keywords = ast.literal_eval(row['Keywords']) if isinstance(row['Keywords'], str) else []
    except:
        keywords = []

    lemmatized = row['Lemmatized_Transcription']
    if isinstance(lemmatized, str):
        try:
            lemmatized = ast.literal_eval(lemmatized)
        except:
            lemmatized = []

    # Remove known irrelevant and already included words
    filtered_words = [
        word for word in lemmatized
        if word not in keywords and word not in irrelevant_words
    ]

    # Prioritize words in the emergency whitelist
    prioritized_words = [word for word in filtered_words if word in emergency_keywords_whitelist]

    # If less than 3 prioritized words, fill with most common others
    count = 3 - len(prioritized_words)
    if count > 0:
        others = [word for word, _ in Counter(filtered_words).most_common() if word not in prioritized_words]
        top_related = prioritized_words + others[:count]
    else:
        top_related = prioritized_words[:3]

    return list(set(keywords + top_related)) #Uses set() to remove any duplicates.

# Apply refined extraction logic
df['extracted_words'] = df.apply(extract_emergency_related_words, axis=1)

# Explode for word-level output
exploded_df = df[['extracted_words', 'Label']].explode('extracted_words').reset_index(drop=True)
exploded_df.columns = ['Word', 'Label']

# Save to CSV
exploded_df.to_csv("semantically_relevant_extracted_words.csv", index=False)
df.head()
```

```python
import pandas as pd
import ast

# Load the dataset
df = pd.read_csv("/content/drive/MyDrive/FYP/final_keywords_emergency_index.csv")

# Define irrelevant/common filler words to exclude
irrelevant_words = set([
    'press', 'one', 'two', 'three', 'four', 'five',
    'ji', 'al', 'alaikum', 'assalam', 'say', 'said',
    'please', 'rescue', 'double', 'call', 'yes', 'no',
    'number', 'mujhy', 'k', 'mein', 'hain', 'ka', 'ki',
    'ke', 'ho', 'raha', 'gaya', 'kar', 'tha', 'hai', 'the',
    'aap', 'ko', 'se', 'for', 'in', 'on', 'is'
])

# Predefined emergency keywords
emergency_keywords = {
    'Accident_Call': {
        'accident', 'crash', 'collision', 'hit', 'strike', 'vehicle', 'wreck',
        'overturn', 'injured', 'bump', 'pileup', 'unconscious', 'motorcycle',
        'car crash', 'road accident', 'bike accident', 'truck collision', 'pedestrian hit',
        'blood', 'road', 'highway', 'driver', 'fatal', 'ambulance', 'ran over', 'traffic', 'damage',
        'victim', 'impact', 'brakes', 'tyre', 'rash driving'
    },

    'Fire_Call': {
        'fire', 'burn', 'flame', 'blaze', 'ignite', 'smoke', 'cylinder', 'explosion',
        'wildfire', 'combustion', 'scorch', 'ash', 'inferno', 'char', 'sparks', 'extinguish', 'alarm',
        'gas leak', 'short circuit', 'electric fire', 'blast', 'firefighter', 'fire brigade',
        'fumes', 'heat', 'boiler', 'chimney', 'toxic smoke', 'electrical fault', 'flammable', 'fire alarm'
    },

    'Medical_Call': {
        'kidney', 'chest', 'abdomen', 'liver', 'heart', 'lung', 'brain', 'stomach', 'appendix',
        'injury', 'unconscious', 'pain', 'bleed', 'medical', 'sick', 'hurt', 'hurting','faint', 'suicide', 'patient','blood',
        'seizure', 'heartattack', 'breathing', 'allergy', 'stroke', 'vomit','vomiting', 'fever', 'diabetic', 'headache',
        'medicine', 'unwell', 'dizzy', 'infection', 'fracture', 'burns', 'bleeding', 'diabetes', 'epilepsy', 'pulse',
        'intensive care', 'ICU', 'coma', 'emergency room', 'nausea', 'bandage'
    }
}


# Flatten and lowercase all emergency keywords
all_emergency_keywords = set(w.lower() for kws in emergency_keywords.values() for w in kws)

# Function to extract relevant keywords
def extract_keywords(row):
    try:
        keywords = ast.literal_eval(row['Keywords']) if isinstance(row['Keywords'], str) else []
    except:
        keywords = []

    try:
        lemmatized = ast.literal_eval(row['Lemmatized_Transcription']) if isinstance(row['Lemmatized_Transcription'], str) else []
    except:
        lemmatized = []

    # Combine both lists
    combined_words = keywords + lemmatized

    # Filtered emergency keywords
    filtered_emergency = [
        word.lower().strip()
        for word in combined_words
        if word.lower().strip() in all_emergency_keywords and word.lower().strip() not in irrelevant_words
    ]

    # If emergency keywords found, use them (limit to 3)
    if filtered_emergency:
        return list(dict.fromkeys(filtered_emergency))[:3]
    else:
        # Else return up to 3 non-irrelevant normal words
        filtered_normal = [
            word.lower().strip()
            for word in combined_words
            if word.lower().strip() not in irrelevant_words
        ]
        return list(dict.fromkeys(filtered_normal))[:3]

# Apply the function
df['Extracted_keywords'] = df.apply(extract_keywords, axis=1)

# Save or preview
df.to_csv("/content/drive/MyDrive/FYP/final_extracted_emergency_keywords.csv", index=False)
```
