```python
import os
import csv
from pydub import AudioSegment
from pydub.playback import play
from noisereduce import reduce_noise
import librosa
import soundfile as sf
import numpy as np
import speech_recognition as sr
from googletrans import Translator
from googletrans import LANGUAGES

# Step 1: Clean and Remove Noise
def clean_audio(input_path, output_path):
    # Load audio file
    y, sr = librosa.load(input_path, sr=None)

    # Reduce noise
    noise_sample = y[:sr]  # Assuming the first second is noise
    reduced_noise = reduce_noise(y=y, y_noise=noise_sample, sr=sr)

    # Save cleaned audio
    sf.write(output_path, reduced_noise, sr)
    print(f"Cleaned audio saved to: {output_path}")

# Step 2: Transcribe Audio using Google Speech Recognition
def transcribe_audio(file_path, language="ur-PK"):
    recognizer = sr.Recognizer()

    # Load audio file
    with sr.AudioFile(file_path) as source:
        audio_data = recognizer.record(source)

    # Transcribe
    try:
        transcription = recognizer.recognize_google(audio_data, language=language)
        print("Transcription complete.")
        return transcription
    except sr.UnknownValueError:
        print("Google Speech Recognition could not understand the audio.")
        return ""
    except sr.RequestError as e:
        print(f"Could not request results from Google Speech Recognition service; {e}")
        return ""

# Step 3: Translate Transcription to English Word by Word
def translate_text_word_by_word(text, source_language="ur", target_language="en"):
    translator = Translator()
    words = text.split()  # Split the transcription into words
    translated_words = []

    for word in words:
        try:
            translation = translator.translate(word, src=source_language, dest=target_language)
            translated_words.append(translation.text)
        except Exception as e:
            print(f"Translation failed for word '{word}': {e}")
            translated_words.append(word)  # Use the original word if translation fails

    return " ".join(translated_words)

if __name__ == "__main__":
    # Directories
    input_folder = "/content/drive/MyDrive/FYP/voices"  # Replace with your folder containing audio files
    output_folder = "/content/drive/MyDrive/FYP/voices cleaned"  # Replace with your folder for cleaned audio files
    csv_output_path = "/content/drive/MyDrive/FYP/transcription 1122"  # Replace with your desired CSV output file
    os.makedirs(output_folder, exist_ok=True)

    # Prepare CSV file
    with open(csv_output_path, mode='w', newline='', encoding='utf-8') as csv_file:
        csv_writer = csv.writer(csv_file)
        csv_writer.writerow(["File Name", "Transcription", "Translation"])

        for file_name in os.listdir(input_folder):
            if file_name.endswith(('.wav', '.mp3')):  # Process only audio files
                input_audio_path = os.path.join(input_folder, file_name)
                cleaned_audio_path = os.path.join(output_folder, f"cleaned_{file_name}")

                # Step 1: Clean the audio
                clean_audio(input_audio_path, cleaned_audio_path)

                # Step 2: Transcribe the cleaned audio
                transcription = transcribe_audio(cleaned_audio_path)
                print(f"Transcription for {file_name}:", transcription)

                # Step 3: Translate the transcription to English word by word
                if transcription:
                    translation = translate_text_word_by_word(transcription, source_language="ur", target_language="en")
                    print(f"Word-by-word Translation for {file_name}:", translation)
                else:
                    translation = ""

                # Write results to CSV
                csv_writer.writerow([file_name, transcription, translation])
                print(f"Results for {file_name} saved to CSV.")
```
