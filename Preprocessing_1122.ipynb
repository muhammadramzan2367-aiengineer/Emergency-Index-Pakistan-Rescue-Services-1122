```python
import pandas as pd
file_path = '/content/drive/MyDrive/FYP/English_Transcription_1122.csv'

# Read the CSV file
df = pd.read_csv(file_path)

# Print column names
print("Column Names:")
print(df.columns.tolist())

# Print number of records
print(f"\nNumber of records: {len(df)}")
```

```python
#cleaned Transcription mean convert all text in lower case:

import re

# Define a function to clean the transcriptions
def clean_text(text):
    # Check if text is a string before applying lower()
    if isinstance(text, str):
        text = text.lower()  # Convert to lowercase
        text = re.sub(r'[^\w\s]', ' ', text)  # Remove punctuation and special characters
        text = re.sub(r'\d+', ' ', text)  # Remove digits
    # If not a string, return original value (or handle it differently)
    # For example, you could convert floats to strings:
    # elif isinstance(text, float):
    #     text = str(text)
    return text



# Apply the cleaning function to the correct column
df['Cleaned_Transcription'] = df['English_Translation'].apply(clean_text)

# Check cleaned transcriptions
df[['English_Translation', 'Cleaned_Transcription']].head()
```

```python
import nltk
from nltk.tokenize import word_tokenize

# Ensure the necessary NLTK packages are downloaded
nltk.download('punkt')
nltk.download('punkt_tab') # Download the punkt_tab data package

# Define a function to handle non-string values
def safe_tokenize(text):
    if isinstance(text, str):
        return word_tokenize(text)
    else:
        # Handle non-string values (e.g., convert to string)
        return word_tokenize(str(text))

# Tokenize the cleaned transcriptions, handling non-string values
df['Tokenized_Transcription'] = df['Cleaned_Transcription'].apply(safe_tokenize)
```

```python
# tokenized sentence : ['This', 'is', 'an', 'emergency', 'call']  Filtered sentence : ['emergency', 'call']

from nltk.corpus import stopwords

# Download NLTK stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Define a function to remove stopwords from tokenized text
def remove_stopwords(tokens):
    return [word for word in tokens if word not in stop_words]

# Apply the stopword removal function
df['Filtered_Transcription'] = df['Tokenized_Transcription'].apply(remove_stopwords)

# Check filtered transcriptions
df[['Tokenized_Transcription', 'Filtered_Transcription']].head()
```

```python
# Filtered: ['running', 'cars'] Lemmatized: ['run', 'car']

from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')

# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

# Define a function to lemmatize the tokens
def lemmatize_tokens(tokens):
    return [lemmatizer.lemmatize(token) for token in tokens]

# Apply lemmatization
df['Lemmatized_Transcription'] = df['Filtered_Transcription'].apply(lemmatize_tokens)

# Check lemmatized transcriptions
df[['Filtered_Transcription', 'Lemmatized_Transcription']].head()
```

```python
# Save the preprocessed data to a new CSV file
output_file = '/content/drive/MyDrive/FYP/preprocessed_transcriptions.csv'
df.to_csv(output_file, index=False)

print("Preprocessed data saved successfully. The file is saved as 'preprocessed_transcriptions.csv'")
```
