```python
import pandas as pd
from googletrans import Translator

# Initialize Translator
translator = Translator()

def translate_urdu_to_english(text):
    try:
        if pd.isna(text):  # Skip NaN values
            return ""
        # Translate full sentence instead of word-by-word
        translated_text = translator.translate(text, src='ur', dest='en').text
        return translated_text
    except Exception as e:
        print(f"Error translating: {text}, Error: {str(e)}")
        return text  # Return original text in case of an error

# **Specify the file path**
file_path = "/content/transcriptions.csv"  # Update this with your actual file path

try:
    # Read CSV File
    df = pd.read_csv("/content/transcriptions.csv")

    # Check if 'Transcription' column exists
    if 'Transcription' in df.columns:
        df['translated_text'] = df['Transcription'].apply(translate_urdu_to_english)

        # Save to a new CSV file
        output_path = file_path.replace(".csv", "_translated.csv")
        df.to_csv(output_path, index=False, encoding='utf-8-sig')  # Ensure correct encoding
        print(f"Translation completed! File saved as: {output_path}")
    else:
        print("Error: 'Transcription' column not found in CSV file.")
except FileNotFoundError:
    print(f"Error: File '{file_path}' not found.")
except Exception as e:
    print(f"Unexpected error: {str(e)}")
```
